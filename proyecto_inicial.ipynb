```python
defunciones= pd.read_csv("defunciones/DEFUNCIONES_FUENTE_DEIS_1990_2021_CIFRAS_OFICIALES.csv",sep=";",encoding="latin-1" )
columnas_filtro = [
    'AÑO',
    'FECHA_DEF',
    'SEXO_NOMBRE',
    'EDAD_TIPO',
    'EDAD_CANT',
    'COD_COMUNA',
    'COMUNA',
    'NOMBRE_REGION',
    'DIAG1',
    'CAPITULO_DIAG1',
    'DIAG2',
    'LUGAR_DEFUNCION'
]
defuncion_nodupli = defunciones.drop_duplicates(subset=columnas_filtro)
defuncion_nodupli.info()
```

```python
lista = []
```

```python
nombres = ["defuncion_nodupli_0_a_4", "defuncion_nodupli_4_a_12", "defuncion_nodupli_12_a_19", "defuncion_nodupli_19_a_40", "defuncion_nodupli_40_a_60", "defuncion_nodupli_60_a_80", "defuncion_nodupli_80_plus"]
```

```python
#edad de 0 a 4 
defuncion_nodupli_0_a_4 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"].isin([4,3,2])) | (defuncion_nodupli["EDAD_CANT"] < 4)]
lista.append(defuncion_nodupli_0_a_4)
defuncion_nodupli_0_a_4
```

```python
#
defuncion_nodupli_4_a_12 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 4) & (defuncion_nodupli["EDAD_CANT"] < 12)]
lista.append(defuncion_nodupli_4_a_12)
defuncion_nodupli_4_a_12
```

```python
defuncion_nodupli_12_a_19 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 12) & (defuncion_nodupli["EDAD_CANT"] < 19)]
lista.append(defuncion_nodupli_12_a_19)
defuncion_nodupli_12_a_19 
```

```python
defuncion_nodupli_19_a_40 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 19) & (defuncion_nodupli["EDAD_CANT"] < 40)]
lista.append(defuncion_nodupli_19_a_40) 
defuncion_nodupli_19_a_40 
```

```python
defuncion_nodupli_40_a_60 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 40) & (defuncion_nodupli["EDAD_CANT"] < 60)]
lista.append(defuncion_nodupli_40_a_60)
defuncion_nodupli_40_a_60
```

```python
defuncion_nodupli_60_a_80 = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 60) & (defuncion_nodupli["EDAD_CANT"] < 80)]
lista.append(defuncion_nodupli_60_a_80)
defuncion_nodupli_60_a_80
```

```python
defuncion_nodupli_80_plus = defuncion_nodupli[(defuncion_nodupli["EDAD_TIPO"] == 1) & (defuncion_nodupli["EDAD_CANT"] >= 80)]
lista.append(defuncion_nodupli_80_plus)
defuncion_nodupli_80_plus
```

```python
lista_tops_edades = []
for i in range (len(lista)):
    df = lista[i].groupby("DIAG1").size().reset_index().sort_values(by=0, ascending=False).reset_index()
    print(nombres[i])
    df.name = nombres[i]
    summon = df.head(5)
    summon.name = nombres[i]
    lista_tops_edades.append(summon)
```

```python
#Es lo mismo pero para diag2
lista_tops_edades_diag_2 = []
for i in range (len(lista)):
    df = lista[i].groupby("DIAG2").size().reset_index().sort_values(by=0, ascending=False).reset_index()
    print(nombres[i])
    df.name = nombres[i]
    summon = df.head(5)
    summon.name = nombres[i]
    lista_tops_edades_diag_2.append(summon)
    
```

```python
for i in lista_tops_edades_diag_2:
    plt.bar(i["DIAG2"], i[0])
    plt.title('EDADES DE '+i.name[-7:] + ' DIAG_2')
    plt.show()
```

```python
#Esto debe rseponder parcialmente la pregunta de las edades
for i in lista_tops_edades:
    plt.bar(i["DIAG1"], i[0])
    plt.title(i.name)
    plt.show()
```
