```python

egresos = pd.read_csv("egresos/EGRESOS_2001/EGRE_DATOS_ABIERTOS_2001.csv", sep=";", encoding="latin-1")
egresos
```

```python
años = range(2001, 2022)
dataframes = {}
for año in años:
    archivo = f"egresos/EGRESOS_{año}/EGRE_DATOS_ABIERTOS_{año}.csv"
    try:
        df = pd.read_csv(archivo, sep=";", encoding="latin-1", low_memory=False)
        dataframes[año] = df 
    except FileNotFoundError:
        print(f"Archivo no encontrado: {archivo}")
```

```python
df_2001 = dataframes[2001]
df_2001
```

```python
df_2001.columns 
```

```python
columns_to_drop = ['ETNIA','GLOSA_PAIS_ORIGEN', 'COMUNA_RESIDENCIA', 'GLOSA_COMUNA_RESIDENCIA','INTERV_Q', 'PROCED']

for year, df in dataframes.items():
    dataframes[year] = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

df_total = pd.concat(dataframes.values(), ignore_index=True)
```

```python
df_total.columns
```

```python
df_total['SEXO'].unique()
```

```python
# Filtrar 'SEXO' para dejar solo 'HOMBRE' y 'MUJER'
df_total = df_total[df_total['SEXO'].isin(['HOMBRE', 'MUJER'])]

# Limpiar 'CONDICION_EGRESO': eliminar '*', convertir '1' y '2' a números
df_total = df_total[df_total['CONDICION_EGRESO'].isin(['1', '2', 1, 2])]
 
# Convertir todos los valores en 'CONDICION_EGRESO' a numéricos
df_total['CONDICION_EGRESO'] = pd.to_numeric(df_total['CONDICION_EGRESO'])
```

```python
df_total['CONDICION_EGRESO'].unique()
```

```python
# Eliminar filas donde 'CONDICION_EGRESO' sea NaN
df_total = df_total.dropna(subset=['CONDICION_EGRESO'])
df_total = df_total.dropna(subset=['DIAG1'])
df_total = df_total.dropna(subset=['DIAG2'])
```

```python
# Convertir 'CONDICION_EGRESO' a categórico con etiquetas descriptivas
df_total['CONDICION_EGRESO'] = df_total['CONDICION_EGRESO'].map({1: 'Vivo', 2: 'Muerto'})

# Verificar que las columnas importantes tengan valores adecuados
print(df_total['CONDICION_EGRESO'].value_counts())
print(df_total['SEXO'].value_counts())
print(df_total['GRUPO_EDAD'].unique())
```

```python
# proporción de mortalidad por género
mortalidad_por_genero = df_total.groupby('SEXO')['CONDICION_EGRESO'].value_counts(normalize=True).unstack() * 100
print(mortalidad_por_genero)

import matplotlib.pyplot as plt

mortalidad_por_genero.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.title('Proporción de Mortalidad por Género')
plt.ylabel('Porcentaje')
plt.xlabel('Género')
plt.legend(title='Condición de Egreso')
plt.show()
```

```python
mortalidad_por_edad = df_total.groupby('GRUPO_EDAD')['CONDICION_EGRESO'].value_counts(normalize=True).unstack() * 100
print(mortalidad_por_edad)

mortalidad_por_edad.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Proporción de Mortalidad por Grupo de Edad')
plt.ylabel('Porcentaje')
plt.xlabel('Grupo de Edad')
plt.legend(title='Condición de Egreso')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Crear tabla cruzada
tabla_cruzada = pd.crosstab(df_total['SEXO'], df_total['CONDICION_EGRESO'])
print(tabla_cruzada)

# Test de Chi-cuadrado
from scipy.stats import chi2_contingency

chi2, p, dof, ex = chi2_contingency(tabla_cruzada)
print(f"Chi-cuadrado: {chi2}, p-valor: {p}")
```

```python
# Proporción de mortalidad por género y grupo de edad
mortalidad_detallada = df_total.groupby(['GRUPO_EDAD', 'SEXO'])['CONDICION_EGRESO'].value_counts(normalize=True).unstack() * 100
print(mortalidad_detallada)

# Gráfico de barras apiladas
mortalidad_detallada.unstack().plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')
plt.title('Proporción de Mortalidad por Género y Grupo de Edad')
plt.ylabel('Porcentaje')
plt.xlabel('Grupo de Edad y Género')
plt.legend(title='Condición de Egreso')
plt.show()
```

```python
# Proporción de mortalidad por región
mortalidad_por_region = df_total.groupby('REGION_RESIDENCIA')['CONDICION_EGRESO'].value_counts(normalize=True).unstack() * 100
print(mortalidad_por_region)

# Gráfico de barras
mortalidad_por_region.plot(kind='bar', stacked=True, colormap='plasma')
plt.title('Proporción de Mortalidad por Región')
plt.ylabel('Porcentaje')
plt.xlabel('Región')
plt.legend(title='Condición de Egreso')
plt.show()
```

```python
import pandas as pd

# Filtrar las columnas necesarias
columnas_relevantes = ['DIAG1', 'DIAG2', 'SEXO', 'CONDICION_EGRESO']
df_filtrado = df[columnas_relevantes].copy()

# Limpiar datos y convertir a texto para evitar problemas de tipo
df_filtrado['DIAG1'] = df_filtrado['DIAG1'].astype(str).str.strip()
df_filtrado['DIAG2'] = df_filtrado['DIAG2'].astype(str).str.strip()

# Contar frecuencia de enfermedades en DIAG1 y DIAG2
enfermedades = pd.concat([df_filtrado['DIAG1'], df_filtrado['DIAG2']])
top_10_enfermedades = enfermedades.value_counts().head(10).index

# Filtrar solo las filas que contienen las enfermedades más comunes
df_top = df_filtrado[
    df_filtrado['DIAG1'].isin(top_10_enfermedades) | 
    df_filtrado['DIAG2'].isin(top_10_enfermedades)
]

# Agrupar por enfermedad, sexo y condición de egreso (vivo/muerto)
estadisticas = df_top.groupby(
    ['DIAG1', 'SEXO', 'CONDICION_EGRESO']
).size().unstack('CONDICION_EGRESO', fill_value=0)

# Mostrar el resultado
print(estadisticas)
```

```python
import matplotlib.pyplot as plt

# Crear un gráfico de barras para las estadísticas
estadisticas.plot(kind='bar', figsize=(12, 6))
plt.title('Estadísticas por género y estado de entrada (Top 10 enfermedades)')
plt.xlabel('Enfermedad')
plt.ylabel('Número de casos')
plt.legend(title='Género y Entrada')
plt.tight_layout()
plt.show()
```

```python
estadisticas_diag2 = df_top.groupby(
    ['DIAG2', 'SEXO', 'CONDICION_EGRESO']
).size().unstack('CONDICION_EGRESO', fill_value=0)

# Mostrar el resultado
print(estadisticas)
```
