```python

años = range(2001, 2022)

# Crear un diccionario para almacenar los DataFrames
dataframes = {}

# Leer archivos y almacenarlos en el diccionario
for año in años:
    archivo = f"egresos/EGRESOS_{año}/EGRE_DATOS_ABIERTOS_{año}.csv"
    try:
        df = pd.read_csv(archivo, sep=";", encoding="latin-1", low_memory=False)
        dataframes[año] = df  # Almacenar DataFrame con el año como clave
    except FileNotFoundError:
        print(f"Archivo no encontrado: {archivo}")

# Si necesitas un DataFrame específico, puedes acceder así:
df_2001 = dataframes[2001]
```
